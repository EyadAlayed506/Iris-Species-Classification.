## My Analysis for the Description Matrix:

*   **Data Integrity**: The minimum value for `Tourists_Number` is zero, so we need to drop any rows where `Tourists_Number` is less than 1.

*   **Distribution**: All numerical columns show a high standard deviation (SD) compared to their mean. This indicates that they are right-skewed, meaning there are more data points concentrated towards the minimum values than towards the maximum values.

Let's compare the average spending per night for Inbound versus Domestic tourism.

```python
#lets visulaize it
import matplotlib.pyplot as plt
import seaborn as sns

# Dynamically select numerical columns
numerical_cols = data.select_dtypes(include=['number']).columns.tolist()

fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=1, figsize=(10, 5 * len(numerical_cols)))

for i, col in enumerate(numerical_cols):
    sns.histplot(data[col], kde=True, ax=axes[i], bins=30)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# fixing the data integrity
data=data[data['Tourists_Number']>=1]
print(data.shape)
# the data may have duplicate rows so we drop it
data_without_duplicates=data.drop_duplicates()
print(f"The number of duplicates rows are {data.shape[0]-data_without_duplicates.shape[0]}")
```

```python
# after dropping the index is not arranged so we reset it
data_without_duplicates.reset_index(inplace=True)
data_without_duplicates=data_without_duplicates.drop("index",axis=1)
```

```python
data_without_duplicates.describe()
```

Let's group the data by `Province`, `YEARS`, and `Tourism_Type` and calculate the mean for all numerical columns. This will help us analyze average tourism trends across different provinces and types over the years.

```python
import pandas as pd

# Group by Province, YEARS, and Tourism_Type and calculate the mean of numerical columns
data_grouped_mean = data_without_duplicates.groupby(['Province', 'YEARS', 'Tourism_Type']).mean().reset_index()

# Display the first few rows of the grouped data
print("Grouped Data (mean):")
(data_grouped_mean.tail())
```

```python
#i wanna see the relation between the number of tourism and the averge speniding

# Get unique provinces
provinces = data_grouped_mean['Province'].unique()

for province in provinces:
    plt.figure(figsize=(12, 7))

    # Filter data for the current province
    df_province = data_grouped_mean[data_grouped_mean['Province'] == province]

    # Plot Domestic Tourism - Tourists Number
    df_domestic = df_province[df_province['Tourism_Type'] == 'Domestic']
    if not df_domestic.empty:
        plt.plot(df_domestic['YEARS'], df_domestic['Tourists_Number'], label='Domestic Tourists Number', marker='o', color='blue')
        # Plot Domestic Tourism - Tourists Spending
        plt.plot(df_domestic['YEARS'], df_domestic['Tourists_Spending'], label='Domestic Tourists Spending', linestyle='--', color='blue')

    # Plot Inbound Tourism - Tourists Number
    df_inbound = df_province[df_province['Tourism_Type'] == 'Inbound']
    if not df_inbound.empty:
        plt.plot(df_inbound['YEARS'], df_inbound['Tourists_Number'], label='Inbound Tourists Number', marker='x', color='red')
        # Plot Inbound Tourism - Tourists Spending
        plt.plot(df_inbound['YEARS'], df_inbound['Tourists_Spending'], label='Inbound Tourists Spending', linestyle='--', color='red')

    plt.title(f'Tourism Trends in {province} Over Years (Tourists Number & Spending)')
    plt.xlabel('Year')
    plt.ylabel('Value (Tourists Number / Spending)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# lets see the correlation for each column


data_grouped_mean[numerical_cols].corr()
```

## My Analysis on the Correlation Matrix

*   **Positive Correlation**: `Tourists_Number`, `Overnight_Stay`, and `Tourists_Spending` are strongly correlated. This is logical, as more tourists generally lead to more nights spent and higher total spending.

*   **Negative Correlation**: We observe a negative correlation between `Tourists_Number` and (`Avg_Spending_Trip`, `Avg_Spending_Night`). This suggests that while total tourist numbers increase, the average spending per trip or per night may decrease. A plausible assumption for this is that high-volume tourism often attracts different traveler segments, such as families or budget-conscious groups, who might prefer more budget-friendly options, thereby reducing the average individual spend even as overall numbers rise.

```python


# Calculate the mean of Avg_Spending_Night for each Province and Tourism_Type
avg_spending_by_province_type = data_grouped_mean.groupby(['Province', 'Tourism_Type'])['Avg_Spending_Night'].mean().reset_index()

plt.figure(figsize=(15, 8)) # Adjust figure size for better readability with many provinces
sns.barplot(x='Province', y='Avg_Spending_Night', hue='Tourism_Type', data=avg_spending_by_province_type, palette='viridis')
plt.title('Average Spending Per Tourist Per Night by Province: Inbound vs. Domestic')
plt.xlabel('Province')
plt.ylabel('Average Spending Per Night')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.legend(title='Tourism Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()

print("Average Spending Per Night by Province and Tourism Type:")
display(avg_spending_by_province_type.head())
```
